```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("../data"+os.sep+"stocks.csv").dropna()
x = data['Low'].values.reshape(-1, 1)
y = data['Close'].values.reshape(-1, 1)
```

```python
# splitting the data into training and test sets, using 20% of the data for test

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=36)
```

```python
%matplotlib inline

plt.scatter(x_train, y_train, s=5)
plt.xlabel('Low price of trading day')
plt.ylabel('Closing price of trading day')
plt.show()
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

model.score(x_train, y_train)
```

```python
predictions = model.predict(x_train)

plt.scatter(x_train, y_train, s=5)
plt.plot(x_train, predictions, color='green')
plt.xlabel('Low price of trading day')
plt.ylabel('Closing price of trading day')
plt.show()
```

```python
predictions = model.predict(x_test)

plt.scatter(x_test, y_test, s=5)
plt.plot(x_test, predictions, color='green')
plt.xlabel('Low price of trading day')
plt.ylabel('Closing price of trading day')
plt.show()
```

```python
mean_squared_error(y_test, predictions)
```
